```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
filepath = 'AmesHousing.csv'
```

```python
data = pd.read_csv(filepath)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data['Yr Sold'],data['SalePrice'],alpha=0.5)
plt.title('Scatter Plot of SalePrice vs Year sold')
plt.xlabel('Year Sold')
plt.ylabel('Sale Price')
plt.grid(True)
```
