```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
filepath = 'AmesHousing.csv'
```

```python
data = pd.read_csv(filepath)
```

```python
data.info()
```

```python
import seaborn as sns


missing_values = data.isnull().mean().sort_values(ascending=False) * 100



print(missing_values.head(20))
```

```python
# Dropping columns with more than 50% missing values
columns_to_drop = ['Pool QC', 'Misc Feature', 'Alley', 'Fence', 'Fireplace Qu', 'Mas Vnr Type']
data_cleaned = data.drop(columns=columns_to_drop)

# Imputing missing values for numerical columns using the median
numerical_cols_with_na = ['Lot Frontage', 'Garage Yr Blt', 'BsmtFin SF 1', 'BsmtFin SF 2', 'Total Bsmt SF']
for col in numerical_cols_with_na:
    data_cleaned[col] = data_cleaned[col].fillna(data_cleaned[col].median())

# Imputing missing values for categorical columns using the mode
categorical_cols_with_na = ['Garage Cond', 'Garage Finish', 'Garage Qual', 'Garage Type', 'Bsmt Exposure', 'BsmtFin Type 1', 'BsmtFin Type 2', 'Bsmt Qual', 'Bsmt Cond']
for col in categorical_cols_with_na:
    data_cleaned[col] = data_cleaned[col].fillna(data_cleaned[col].mode()[0])
```
